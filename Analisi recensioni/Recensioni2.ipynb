```python
import pandas as pd
import re

# Funzione per pulire il testo della recensione
def pulisci_testo(testo):
    # Controlla se il testo √® una stringa
    if isinstance(testo, str):
        # Rimozione di URL
        testo = re.sub(r'http\S+', '', testo)
        # Rimozione di caratteri speciali e numeri
        testo = re.sub(r'[^a-zA-Z\s]', '', testo, re.I|re.A)
        # Conversione in minuscolo
        testo = testo.lower()
        # Rimozione degli spazi extra
        testo = testo.strip()
        return testo
    else:
        # Se non √® una stringa, ritorna un testo vuoto o un valore predefinito
        return ""

# Applicazione della funzione di pulizia al DataFrame
df['reviewDescription'] = df['reviewDescription'].apply(pulisci_testo)
```

```python
import pandas as pd
df = pd.read_csv('Recensioni_new_label.csv')
```

```python
sentimenti = [
    "Agisce bene",
    "Lascia macchie",
    "Fedelt√† alla marca",
    "Cambier√≤ marca",
    "Buon profumo",
    "Odore sgradevole",
    "Buone ricariche",
    "Brutte ricariche",
    "Buon prezzo",
    "Troppo costoso",
    'Nulla'
]
```

```python
import pandas as pd
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, precision_recall_curve
import numpy as np

#rinomina la colonna sentiment_labels in sentiment_label
df.rename(columns={'sentiment_labels': 'sentiment_label'}, inplace=True)

df = df[df['sentiment_label'].notnull()]
df = df[df['sentiment_label'] != 'nan']
print(df.shape)

# Modifica della funzione per normalizzare i label
def normalize_label(label):
    if isinstance(label, float):  # Controlla se il label √® un float (NaN)
        return label
    label = label.lower().strip()
    return label

# Normalizzazione della lista dei sentimenti
sentimenti = [normalize_label(sentimento) for sentimento in sentimenti]

# Modifica della funzione per pulire e normalizzare le etichette del dataset
def clean_and_normalize_labels(label):
    if pd.isna(label):
        return label
    label = str(label)
    label = label.strip("[]")
    label = label.replace('"', '').replace("'", "")
    label = normalize_label(label)
    return label

# Modifica della funzione di mappatura per gestire i label normalizzati
def map_labels_to_sentiments(label_list):
    mapped_labels = []
    for label in label_list:
        normalized_label = normalize_label(label)
        print(f"Elaborazione: {normalized_label}")  # Stampa per il debugging
        if normalized_label in sentimenti:
            mapped_labels.append(normalized_label)
            print(f"Mappato: {normalized_label}")
        else:
            print(f"Ignorato: {normalized_label}")
    return mapped_labels

# Applicazione delle funzioni modificate
df['sentiment_label'] = df['sentiment_label'].astype(str).apply(clean_and_normalize_labels)
df['sentiment_label'] = df['sentiment_label'].apply(lambda x: x.split(','))
df['sentiment_label'] = df['sentiment_label'].apply(map_labels_to_sentiments)

# Normalizzazione dei label unici nel dataset
unique_labels_in_dataset = set(df['sentiment_label'].explode().apply(normalize_label).unique())

# Calcolo del numero di valori in sentimenti che non sono nel dataset
missing_values_count = len(set(sentimenti) - unique_labels_in_dataset)
print(f"Numero di valori di sentimenti non presenti nel dataset: {missing_values_count}")

# Stampa dei valori che non sono nel dataset
missing_values = set(sentimenti) - unique_labels_in_dataset
print(f"Valori di sentimenti non presenti nel dataset: {missing_values}")
```

```python
# Suddivisione in train, validation e test set
train_val_df, test_df = train_test_split(df, test_size=0.15, random_state=21)
train_df, val_df = train_test_split(train_val_df, test_size=0.20, random_state=21)

# Tokenizzazione
model_name = "indigo-ai/BERTino"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)

train_encodings = tokenizer(list(train_df['reviewDescription']), truncation=True, padding=True, max_length=512)
val_encodings = tokenizer(list(val_df['reviewDescription']), truncation=True, padding=True, max_length=512)
test_encodings = tokenizer(list(test_df['reviewDescription']), truncation=True, padding=True, max_length=512)

# Creazione del MultiLabelBinarizer basato sui sentimenti
mlb = MultiLabelBinarizer(classes=sentimenti)
train_labels = mlb.fit_transform(train_df['sentiment_label'])
val_labels = mlb.transform(val_df['sentiment_label'])
test_labels = mlb.transform(test_df['sentiment_label'])

# Definizione della classe Dataset
class AmazonReviewDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.float)
        return item

    def __len__(self):
        return len(self.labels)

# Crea i dataset
train_dataset = AmazonReviewDataset(train_encodings, train_labels)
val_dataset = AmazonReviewDataset(val_encodings, val_labels)
test_dataset = AmazonReviewDataset(test_encodings, test_labels)

# Carica il modello
model = DistilBertForSequenceClassification.from_pretrained(model_name, 
                                                            num_labels=len(mlb.classes_),
                                                            problem_type="multi_label_classification")

# Parametri di training senza early stopping
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=50,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=20,
    weight_decay=0.1,
    learning_rate=5e-5,
    logging_dir='./logs',
    logging_steps=25,
    evaluation_strategy="steps",
    save_strategy="steps",
    save_steps= 1500,
    load_best_model_at_end=True
)
```

```python
# Funzione per calcolare le metriche
def compute_metrics(pred):
    logits = pred.predictions[0]
    preds = torch.sigmoid(torch.tensor(logits)).numpy() > 0.17
    labels = pred.label_ids
    precision = precision_score(labels, preds, average='micro')
    recall = recall_score(labels, preds, average='micro')
    f1 = f1_score(labels, preds, average='micro')
    return {
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments, TrainerCallback

# Callback per Early Stopping
class EarlyStoppingCallback(TrainerCallback):
    def __init__(self, early_stopping_patience: int):
        self.early_stopping_patience = early_stopping_patience
        self.best_f1 = None
        self.patience_counter = 0

    def on_evaluate(self, args, state, control, metrics=None, **kwargs):
        if metrics is None:
            return

        eval_f1 = metrics.get("eval_f1")

        if self.best_f1 is None or eval_f1 > self.best_f1:
            self.best_f1 = eval_f1
            self.patience_counter = 0
        else:
            self.patience_counter += 1

        if self.patience_counter >= self.early_stopping_patience:
            print(f"Early stopping triggered at step {state.global_step}")
            control.should_training_stop = True

# Inizializzazione del Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)]
)

# Avvia il training
trainer.train()
```

```python
import re

# Funzione per normalizzare i valori di sentiment_label
def normalize_sentiment_label(label):
    if pd.isna(label):
        return []
    if isinstance(label, list):
        # Se √® gi√† una lista, normalizza ogni elemento
        return [s.lower().strip() for s in label]
    if isinstance(label, str):
        # Rimuove parentesi quadre e virgolette e gestisce la stringa
        label = re.sub(r"[\[\]']", '', label)
        return [s.strip().lower() for s in label.split(';')]
    # Se il tipo di dati non √® n√© una stringa n√© una lista
    return []

df['sentiment_label'] = df['sentiment_label'].apply(normalize_sentiment_label)

# Assicurati che anche la lista sentimenti sia in minuscolo
sentimenti = [s.lower() for s in sentimenti]

# Creazione di colonne separate per ogni sentimento
for sentimento in sentimenti:
    df[sentimento] = df['sentiment_label'].apply(lambda labels: 1 if sentimento in labels else 0)
```
